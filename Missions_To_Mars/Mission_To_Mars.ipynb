```python
#Dependencies

from bs4 import BeautifulSoup
import os
import requests
import pymongo
import lxml
```

```python
# NASA Mars News

# Creating Option to Read HTML - URL option did not show information that needed to be pulled.
# When soup was printed using prettify method, the title and paragraph for the lastest article was absent from view.

filepath = os.path.join("Resources", "mission_to_mars.html")
with open(filepath) as file:
    html = file.read()
```

```python
# Create BeautifulSoup object 

soup = BeautifulSoup(html, 'html.parser')
type(soup)
```

```python
print(soup.prettify())
```

```python
# Return Section List

result = soup.body.find_all('section', class_ = 'image_and_description_container')
print(result)
```

```python
# Collect the latest News Title and Paragraph Text 

for data in result:
    lastest_title = data.find('div', class_='content_title').text
    latest_p = data.find('div', class_='article_teaser_body').text
    print(lastest_title)
    print(latest_p)
```
